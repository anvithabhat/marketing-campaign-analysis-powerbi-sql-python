```python
#check for existence of null values
df.isnull().sum()
```

```python
#change column names to all lower case
df.columns=df.columns.str.lower()
```

```python
df.duplicated(subset='campaign id')
```

```python
df.columns
```

```python
#replace space in column name with _
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
#replacing null values in click, conversion and impressions with MEAN values of respective column

fill_values={'impressions':df['impressions'].mean(),
             'clicks':df['clicks'].mean(),
            'conversions':df['conversions'].mean()}
#reassign fill_values to column 
df=df.fillna(value=fill_values)
```

```python
#for handelling null in categorical value column i.e., channel, used MODE
#for handelling missing cose and revenue values, used MEDIAN
fill_values1={'channel':df['channel'].mode()[0],
            'cost':df['cost'].median(),
            'revenue':df['revenue'].median()}
#reassign fill_values to column 
df=df.fillna(value=fill_values1)
```

```python
df
```

```python
df['customers'].isnull().sum()
```

```python
df['customers'] = df['customers'].fillna(df['conversions'])
```

```python
df
```

```python
df['customers']=df['customers'].astype(int)
```

```python
df
```

```python
df[['impressions','clicks','conversions']]=df[['impressions','clicks','conversions']].astype(int)
```

```python
df
```

```python
#Feature engineering
#to add Click through rate(CTR), it measures engagement
df['ctr']=df['clicks']/df['impressions']

#conversion rate, meassures effectiveness after clicks
df['conversion_rate']=df['conversions']/df['clicks']

#Cost per customer(CPC), how much to pay per click
df['cac']=df['cost']/df['customers']

#Profit
df['profit']=df['revenue']-df['cost']
```

```python
df
```

```python
df[['ctr','conversion_rate','cac','profit']].dtypes
```

```python
df[['ctr','conversion_rate','cac','profit']]=df[['ctr','conversion_rate','cac','profit']].round(2)
```
